# Forward Propagation using Dot Product

```python
input_data = np.array([2,3])
weights = {'node_0':np.array([1,1]),
           'node_1':np.array([-1,1]),
           'output':np.array([2,-1])}
node_0_value = (input_data*weights['node_0']).sum()
node_1_value = (input_data*weights['node_1']).sum()
```

```python
hidden_layer_values = np.array([node_0_value, node_1_value])
print(hidden_layer_values)
```

```python
output = (hidden_layer_values * weights['output']).sum()
print(output)
```

# **Example Dot Product**
With our own example, we decided to create neural network with 3 input nodes, 3 hidden layer nodes, and 1 output node for simplicity. In a more realistic model, there will be an abundance of nodes in all three layers depending on the objective. we started with 1, 2, 3 as our input nodes. with weights of -1, 1, 2 on input node 1, 1, -1, 3 for input node 2, and 2, 3, 4 for input node 3.

**MATH:**


> hidden layer node 1: (1**-1) + (2**1) + (3*2) = 7

> hidden layer node 2: (1**1) + (2**-1) + (3*3) = 8

> hidden layer node 3: (1**2) + (2**3) + (3*4) = 20

> output layer = 58
